```python
def read_dataframe(url):
    df = pd.read_parquet(url)
    df["duration"] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda duration: duration.total_seconds() / 60)
    df = df[(df.duration >= 1) & (df.duration <= 60)]
    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)

    return df
```

```python
# df_validate = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
df_validate = read_dataframe('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
df_validate
```

```python
numerical = ['duration']
```

```python
val_dicts = df_validate[categorical + numerical].to_dict(orient='records')
```

```python
X_val = dv.transform(val_dicts)
```

```python
y_val = df_validate['duration'].values
```

```python
y_pred = lr.predict(X_val)
```

```python
root_mean_squared_error(y_val, y_pred)
```
